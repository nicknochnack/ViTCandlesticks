```python
from PIL import Image 
```

```python
def balance(label:int, examples:int) -> None: 
    df_slice = df[df['Label']==label]
    res = np.random.choice(len(df_slice['Label']), examples)

    for idx in res: 
        img_name = df_slice['Image'].iloc[idx]
        img = Image.open(f'consolidated_data/{img_name}')
        new_image_name = uuid.uuid1()
        img.save(f'consolidated_data/{new_image_name}.png')
        data_list.append(np.array([f'{new_image_name}.png', int(label)], dtype='object'))
        
    return res
    
balance(1, 1)
```
